```python
def rush_hour(time):
    if ((time >= 7 and time <= 9) or (time >= 15 and time <= 18)):
        return 1
    else:
        return 0
def rush_hour_day(day):
    if (day >= 0 and day <= 4):
        return 1
    else:
        return 0

df['rush_hour'] = (df['datetime'].apply(rush_hour) & df['day_of_week'].apply(rush_hour_day))
df['rush_hour'].value_counts()
```
