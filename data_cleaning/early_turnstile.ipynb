```python
import pandas as pd
import pickle as pkl
import warnings
```

```python
d2010 = pd.read_csv('get_data/turnstile/turnstiles/pre_oct_18/2010.csv')
d2011 = pd.read_csv('get_data/turnstile/turnstiles/pre_oct_18/2011.csv')
d2012 = pd.read_csv('get_data/turnstile/turnstiles/pre_oct_18/2012.csv', error_bad_lines=False)
d2013 = pd.read_csv('get_data/turnstile/turnstiles/pre_oct_18/2013.csv')
d2014 = pd.read_csv('get_data/turnstile/turnstiles/pre_oct_18/2014_pre_oct_18_excl.csv')
warnings.filterwarnings('ignore')
```

```python
stuff = "C/A,UNIT,SCP,DATE_1,TIME_1,DESC_1,ENTRIES_1,EXITS_1,DATE_2,TIME_2,DESC_2,ENTRIES_2,EXITS_2,DATE_3,TIME_3,DESC_3,ENTRIES_3,EXITS_3,DATE_4,TIME_4,DESC_4,ENTRIES_4,EXITS_4,DATE_5,TIME_5,DESC_5,ENTRIES_5,EXITS_5,DATE_6,TIME_6,DESC_6,ENTRIES_6,EXITS_6,DATE_7,TIME_7,DESC_7,ENTRIES_7,EXITS_7,DATE_8,TIME_8,DESC_8,ENTRIES_8,EXITS_8"
```

```python
cols = list(stuff.split(','))
```

```python
cols
```

```python
df2010 = pd.DataFrame(d2010)
df2011 = pd.DataFrame(d2011)
df2012 = pd.DataFrame(d2012)
df2013 = pd.DataFrame(d2013)
df2014 = pd.DataFrame(d2014)
```

```python
pd.set_option('display.max_columns',None)
```

```python
df2010.columns = cols
df2011.columns = cols
df2012.columns = cols
df2013.columns = cols
df2014.columns = cols
```

```python
df2010.columns = df2010.columns.str.strip().str.lower()
df2011.columns = df2011.columns.str.strip().str.lower()
df2012.columns = df2012.columns.str.strip().str.lower()
df2013.columns = df2013.columns.str.strip().str.lower()
df2014.columns = df2014.columns.str.strip().str.lower()
```

```python
all_dfs = pd.concat([df2010, df2011, df2012, df2013, df2014])
```

```python
all_dfs
```

```python
with open('pre_oct_2014.pkl', 'wb') as f:
    pkl.dump(all_dfs, f)
```
