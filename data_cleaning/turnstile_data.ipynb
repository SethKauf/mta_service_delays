```python
ent = pd.read_csv('data/turnstile/entries.csv')
ext = pd.read_csv('data/turnstile/exits.csv')
```

```python
ent.set_index('date',inplace=True)
ext.set_index('date',inplace=True)
```

```python
cols_en = sorted(list(ent.columns))
cols_ex = sorted(list(ext.columns))
```

```python
ent = ent.reindex(columns=cols_en)
ext = ext.reindex(columns=cols_ex)
```
