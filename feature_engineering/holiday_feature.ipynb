```python
import pandas as pd
import pickle as pkl
import datetime as dt
pd.set_option('display.max_columns',None)
```

```python
with open('../data/full_data.pkl','rb') as f:
    df = pkl.load(f)
```

```python
from pandas.tseries.holiday import USFederalHolidayCalendar
cal = USFederalHolidayCalendar()
holidays = cal.holidays(start='2010-01-01', end='2021-12-31').to_pydatetime()
# if datetime.datetime(2014,01,01) in holidays:
#     print True 
```

```python
def get_one(value):
    if value is True:
        return 1
    else:
        return 0
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['just_date'] = df['date'].dt.date
df['just_date'] = pd.to_datetime(df['just_date'])
```

```python
holidays = pd.Series(holidays)
```

```python
df.insert(455,'holiday',0)
df['holiday'] = df['just_date'].isin(holidays)
df['holiday'] = df['holiday'].apply(get_one)
df.drop('just_date',axis=1,inplace=True)
```
