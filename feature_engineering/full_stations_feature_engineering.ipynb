```python
def become_one(one):
    if one >= 1:
        return 1
    else:
        return 0
    
def get_one(value):
    if value is True:
        return 1
    else:
        return 0
```

# Creating Features

```python
causes_list = ['door','unsanitary','mechanical','power','inspection','impact','accident','derail',
              'maintenance','brake','track','signal','switch','fire','smoke','construction','fdny',
              'block','station','flood','falling','gap','debris','tree',' rail','electric',
              'sick','medical','police','nypd','investigation','ems','struck','assault','graffiti',
              'surf','vandal','storm','inclement','flood','ice','cold',' rain','icy','forecast','hurricane',
              'on track']

direction_list = ['northbound','soundbound','both direction','nb','sb','bd','bothdirections','both way','bothway','n /','n/b',
            'n/ b','s /','s/b','s/ b','b/d','b / d','n / b','s / b','b/ d','b /d','b/ ','manhattan bound', 'brooklyn bound',
            'queens bound', 'bronx bound','manh bound','bk bound','bx bound','qns bound']
```

```python
df['causes'] = df['message_lower'].str.extract('({0})'.format('|'.join(causes_list)))
```

```python
causes = pd.get_dummies(df['causes'])
df = df.merge(causes, how='outer',left_index=True,right_index=True)
```

```python
df['passenger_disruption'] = (df['medical'] + df['sick'] + df['police'] + df['graffiti']
                                         + df['vandal'] + df['investigation'] + df['ems'] + df['nypd']
                                         + df['struck'] + df['assault'] + df['surf'])
df['passenger_disruption'].apply(become_one)

df['station_integrity_disruption'] = (df['station'] + df['construction'] + df['unsanitary'] + df['inspection'] + df['impact']
                                      + df['flood'] + df['gap'] + df['smoke'] + df['fdny'] )
df['station_integrity_disruption'].apply(become_one)

df['mechanical_and_track_disruption'] = (df['mechanical'] + df['track'] + df['derail'] + df[' rail']
                                         + df['debris'] + df['power'] + df['electric'] + df['fire']
                                         + df['maintenance'] + df['block'] + df['falling'] + df['on track'] + df['tree'])
df['mechanical_and_track_disruption'].apply(become_one)

df['train_disruption'] = (df['door'] + df['accident'] + df['brake'])
df['train_disruption'].apply(become_one)

df['weather_disruption'] = (df['storm'] + df['inclement'] + df['flood'] + df['ice']
                            + df['cold'] + df[' rain'] + df['icy'] + df['forecast'] + df['hurricane'])
df['weather_disruption'].apply(become_one)

df['signal_disruption'] = df['signal']

df['switch_disruption'] = df['switch']
```

```python
df.drop(['station','construction','unsanitary','inspection','impact','mechanical',
        'track','derail',' rail','debris','power','electric','sick','medical','police',
        'nypd','investigation','ems','struck','assault','graffiti','surf','vandal', 'block',
        'gap','flood','smoke','fire','door', 'accident', 'maintenance', 'brake','falling','on track',
        'tree', 'storm', 'inclement', 'flood', 'ice', 'cold', ' rain', 'icy', 'forecast', 'hurricane',
        'signal','switch','causes','fdny'],axis=1,inplace=True)
```

```python
df['direction_msg'] = df['message_lower'].str.extract('({0})'.format('|'.join(direction_list)))

df['direction_sub'] = df['subj_lower'].str.extract('({0})'.format('|'.join(direction_list)))

direction_1 = pd.get_dummies(df['direction_msg'])
direction_2 = pd.get_dummies(df['direction_sub'])

direction = direction_1.merge(direction_2,left_index=True,right_index=True)
```

```python
direction['direction_bronx'] = (direction['bronx bound'] + direction['bx bound'])
direction['direction_bronx'] = direction['direction_bronx'].apply(become_one)

direction['direction_brooklyn'] = (direction['brooklyn bound_x'] + direction['bk bound']
                                  + direction['brooklyn bound_y'])
direction['direction_brooklyn'] = direction['direction_brooklyn'].apply(become_one)

direction['direction_manhattan'] = (direction['manhattan bound'] + direction['manh bound'])
direction['direction_manhattan'] = direction['direction_manhattan'].apply(become_one)

direction['direction_queens'] = (direction['qns bound_x'] + direction['qns bound_y'] + direction['queens bound'])
direction['direction_queens'] = direction['direction_queens'].apply(become_one)

# both/unspec directions
direction['direction_both'] = (direction['b/ '] + direction['b/ d'] + direction['bd']
                               + direction['both direction_x'] + direction['both direction_y'])
direction['direction_both'] = direction['direction_both'].apply(become_one)

direction['direction_north'] = (direction['n /_x'] + direction['n/ b'] + direction['n/b']
                               + direction['nb'] + direction['northbound_x'] + direction['n /_y']
                               + direction['northbound_y'])
direction['direction_north'] = direction['direction_north'].apply(become_one)

direction['direction_south'] = (direction['s /_x'] + direction['s/ b'] + direction['s/b']
                                + direction['sb_x'] + direction['s /_y'] + direction['sb_y'])
direction['direction_south'] = direction['direction_north'].apply(become_one)
```

```python
direction.drop(['b/ ', 'b/ d', 'bd', 'both direction_x', 'bronx bound', 'brooklyn bound_x', 'manhattan bound',
                'n /_x', 'n/ b', 'n/b', 'nb', 'northbound_x', 'qns bound_x', 's /_x', 's/ b', 's/b',
                'sb_x', 'bk bound', 'both direction_y', 'brooklyn bound_y','bx bound', 'manh bound', 'n /_y', 'northbound_y',
                'qns bound_y', 's /_y', 'sb_y', 'queens bound'],axis=1,inplace=True)
direction
```

```python
df = df.merge(direction, how='left', left_index=True,right_index=True)
df.drop(['direction_msg','direction_sub'],axis=1,inplace=True)
```

```python
df['major_conditions'] = df['major_conditions'].apply(become_one)
df['minor_conditions'] = df['minor_conditions'].apply(become_one)
df['neutral_conditions'] = df['neutral_conditions'].apply(become_one)
df['clear_conditions'] = df['clear_conditions'].apply(become_one)
```

```python
df['max_temp'].ffill(inplace=True)
df['avg_temp'].ffill(inplace=True)
df['min_temp'].ffill(inplace=True)
df['prcp'].ffill(inplace=True)
df['snowfall'].ffill(inplace=True)
df['snowfall_depth'].ffill(inplace=True)
df['fastest_wind_spd'].ffill(inplace=True)
df['avg_wind_spd'].ffill(inplace=True)
df['fastest_wind_directions'].ffill(inplace=True)
```

```python
df['bronx'] = df['subject'].str.contains('bx')
df['manhattan'] = df['subject'].str.contains('manh')
df['brooklyn'] = df['subject'].str.contains('bk')
df['queens'] = df['subject'].str.contains('qns')

df['brooklyn'] = df['brooklyn'].apply(get_one)
df['queens'] = df['queens'].apply(get_one)
df['bronx'] = df['bronx'].apply(get_one)
df['manhattan'] = df['manhattan'].apply(get_one)
```

```python
col_order = ['date', 'subject','subj_lower', 'message', 'message_lower','delayed_div_a', 'delayed_div_b',
             'bronx','brooklyn','manhattan','queens', 'direction_bronx','direction_brooklyn',
             'direction_manhattan', 'direction_queens','direction_both','direction_north', 'direction_south',
             'passenger_disruption', 'station_integrity_disruption',
             'mechanical_and_track_disruption', 'train_disruption', 'weather_disruption', 'signal_disruption',
             'switch_disruption', 'mon', 'tue', 'wed', 'th', 'fri', 'weekend','rush_hour',
             'max_temp','avg_temp', 'min_temp','avg_wind_spd','fastest_wind_directions', 'fastest_wind_spd',
             'major_conditions','minor_conditions', 'neutral_conditions', 'clear_conditions', 'prcp',
             'snowfall', 'snowfall_depth','delayed_1', 'delayed_2', 'delayed_3', 'delayed_4', 'delayed_42', 'delayed_5',
              'delayed_6', 'delayed_7', 'delayed_a', 'delayed_b', 'delayed_c', 'delayed_d',  'delayed_e', 'delayed_f',
             'delayed_frnk', 'delayed_g', 'delayed_j', 'delayed_l', 'delayed_m', 'delayed_n', 'delayed_q', 'delayed_r',
             'delayed_rck', 'delayed_w', 'delayed_z']
```
