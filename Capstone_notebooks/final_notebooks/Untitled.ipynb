```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
iris.data
```

```python
iris.target
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
KM_model = kmeans.fit(iris.data)
```

```python
KM_model.labels_
```

```python
KM_model.cluster_centers_
```

```python
import pandas as pd

pd.crosstab(iris.target, KM_model.labels_)
```
