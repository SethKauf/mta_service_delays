# get date and time from messages
def get_dates(messages):
    dates = []
    times = []
    for i in range(len(messages)):
        date = messages[i].text.strip()[7:14]
        time = messages[i].text.strip()[14:21]
        dates.append(date)
        times.append(time)
        i += 1
    return pd.DataFrame(zip(dates, times), columns=['date', 'time'])
